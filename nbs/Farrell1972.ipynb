# Farrell 1972 [Exposed]
The purpose of this notebook is to work through Farrel (1972) and show that computation of elastic Love numbers is not just a task that should be "left to specialists" (Bevis et. al 2016). In fact, with a little theory, and some numerical chops, we can see exactly where they come from.

$$\nabla\cdot\tau - \nabla(\rho g \mathbf{s}\cdot\mathbf{e_r}) - \rho\nabla\phi + g\nabla\cdot(\rho\mathbf{s})\mathbf{e_r}-\omega^2\rho\mathbf{s} = 0$$

$$\nabla^2\phi = -4\pi G\nabla\cdot(\rho\mathbf{s})$$

$\phi=\phi_1+\phi_2$, where $\phi_1$ is the perturbation potential associated with the load, and $\phi_2$ is any external applied potential.

The key to solving these equations for a spherical Earth is to expand all of these quantities into the spherical harmonics. In the symmetric cases considered here, we need only consider the expansion into the Legendre polynomials.

## Establish Boundary Conditions
In determining how the Earth responds, we need to decide what it's responding to. We could choose to construct a disk load, which is a fairly common technique for problems in the geophysical context, and which illucidates interesting features of the equations of motion. Alternatively, we could decide that to break down our load into a series of point loads or unit harmonic loads and construct solutions be superposition of these responses.

$$\gamma_{\textrm{pointload}}= \sum_n \frac{2n+1}{2\pi a^2}P_n(\cos\theta)$$
$$\gamma_{\textrm{harmonic, n}} = P_n(\cos\theta)$$

We can, however, generalize these loads to

$$\gamma = \sum_{n,m} C_n^m Y_n^m(\theta,\phi).$$

This load is a radial traction, $\mathbf{t}(r_e, \theta,\phi)=-g(r_e)\gamma(\theta,\phi)\mathbf{\hat{r}} + 0\mathbf{\hat{\theta}} + 0\mathbf{\hat{\phi}}$, as well as a perturbation to the gravitational potential, $Q_2(r_e)=-4\pi G C_n^m$. Expanding these expressions into spherical harmonics yields the following boundary conditions at the surface.

$$\left[\begin{array}{c}
~\\ ~\\T_L(r_e)\\T_M(r_e)\\ ~\\ Q_2(r_e) 
\end{array}\right] = \left[\begin{array}{c}

\end{array}\right]$$

$$\ell$$

$$\varepsilon_{rr}$$

Expanding the displcements into the vector spherical harmonic basis,
$$
\mathbf{u}(\mathbf{r}) = \sum_{\ell,m}\left[U_L(r)\mathbf{L}(\theta,\phi) + U_M(r)\mathbf{M}(\theta,\phi)+U_N(r)\mathbf{N}(\theta,\phi)\right]$$
which can re-express in the spherical basis
$$\mathbf{u}(\mathbf{r}) = \sum_{\ell,m}\left[ U_L(r)Y_\ell^m(\theta,\phi)\mathbf{\hat{r}} + \left(U_M(r)\frac{\partial Y_\ell^m(\theta,\phi)}{\partial\theta} + \frac{U_N(r)}{\sin\theta}\frac{\partial Y_\ell^m(\theta,\phi)}{\partial \phi}\right)\mathbf{\hat{\theta}} + \left(\frac{U_M(r)}{\sin\theta}\frac{\partial Y_\ell^m(\theta,\phi)}{\partial \phi} - U_N(r)\frac{\partial Y_\ell^m(\theta,\phi)}{\partial \theta}\right)\mathbf{\hat{\phi}}\right].$$

From here on in, we'll suppress the arguments of the function (radial dependence is packaged into the expansion coefficients and angular into the spherical harmonics). The approach will be to express all the equations of state and motion with spherical harmonic expansions in the spherical basis, and compare similar terms to obtain the relevant equations in the vector spherical harmonic basis. I'll illustrate this first for the more straightforward case of the stress-strain relationships.

### Stress-Strain Constitutive Equations
Consider first $\tau_{rr}$. In the spherical basis it is written
$$\tau_{rr} = (\lambda + 2\mu)\varepsilon_{rr} + \lambda\varepsilon_{\theta\theta}+\lambda\varepsilon_{\phi\phi},$$
and in the vector spherical harmonic expansion,
$$\tau_{rr} = \sum_{\ell,m} T_L Y_\ell^m.$$
Noting that 
$$\varepsilon_{rr} = \frac{\partial u_r}{\partial r} = \frac{\partial}{\partial r}\sum_{\ell,m}U_LY_\ell^m=\sum_{\ell,m}\frac{\partial U_L}{\partial r}Y_\ell^m$$
and
$$\begin{align}
r(\varepsilon_{\theta\theta}+\varepsilon_{\phi\phi}) &= 2 u_r + \frac{\partial u_\theta}{\partial \theta} + \cot\theta u_\theta + \frac{1}{\sin\theta}\frac{\partial u_\phi}{\partial \phi}\\
&= \sum_{\ell,m}\left[ 2U_L Y_\ell^m + \frac{\partial}{\partial \theta} \left(U_M\frac{\partial Y_\ell^m}{\partial\theta} + \frac{U_N}{\sin\theta}\frac{\partial Y_\ell^m}{\partial \phi}\right) + \cot\theta\left(U_M\frac{\partial Y_\ell^m}{\partial\theta} + \frac{U_N}{\sin\theta}\frac{\partial Y_\ell^m}{\partial \phi}\right) + \frac{1}{\sin\theta}\frac{\partial}{\partial\phi}\left(\frac{U_M}{\sin\theta}\frac{\partial Y_\ell^m}{\partial \phi} - U_N\frac{\partial Y_\ell^m}{\partial \theta}\right)\right]\\
&= \sum_{\ell,m}\left[ 2U_L Y_\ell^m + U_M\left(\frac{\partial^2 Y_\ell^m}{\partial\theta^2} + \cot\theta\frac{\partial Y_\ell^m}{\partial\theta} + \frac{1}{\sin^2\theta}\frac{\partial^2 Y_\ell^m}{\partial \phi^2}\right)\right]\\
&= \sum_{\ell,m}\left[ 2U_L Y_\ell^m + r^2\nabla_S^2Y_\ell^m\right]\\
&= \sum_{\ell,m}\left[ 2U_L  - \ell(\ell+1)U_M\right]Y_\ell^m.
\end{align}$$
Comparing terms in the spherical harmonic expansions, it follows that
$$T_L = (\lambda + 2\mu)\frac{\partial U_L}{\partial r} + \frac{2\lambda}{r}U_L  - \frac{\ell(\ell+1)\lambda}{r}U_M,$$
which is, incidentally, a first order differential equation for $U_L$:
$$\frac{\partial U_L}{\partial r} =  - \frac{2\lambda}{r(\lambda + 2\mu)}U_L  + \frac{\ell(\ell+1)\lambda}{r(\lambda + 2\mu)}U_M+ \frac{T_L}{\lambda + 2\mu}$$

Next, let's exapnd $\tau_{r\theta}$. Again, in the spherical basis,
$$\tau_{r\theta}=2\mu\varepsilon_{r\theta}=\mu\frac{\partial u_\theta}{\partial r} + \frac{\mu}{r}\left(\frac{\partial u_r}{\partial \theta} - u_\theta\right).$$
And in the vector spherical harmonic basis,
$$\tau_{r\theta}=\sum_{\ell,m}\left(T_M\frac{\partial Y_\ell^m}{\partial \theta} + \frac{T_N}{\sin\theta}\frac{Y_\ell^m}{\partial\phi}\right).$$
Expanding the right-hand side of the strain (using the expansions of $u_r$ and $u_\theta$ from above),
$$\begin{align}
2\mu\varepsilon_{r\theta}&=\mu\sum_{\ell,m}\left[\frac{\partial U_M}{\partial r}\frac{\partial Y_\ell^m}{\partial\theta} + \frac{\partial U_N}{\partial r}\frac{1}{\sin\theta}\frac{\partial Y_\ell^m}{\partial\phi}+\frac{U_L}{r}\frac{\partial Y_\ell^m}{\partial\theta} - \frac{1}{r}\left(U_M\frac{\partial Y_\ell^m}{\partial\theta}+\frac{U_N}{\sin\theta}\frac{\partial Y_\ell^m}{\partial\phi}\right)\right]\\
&=\mu\sum_{\ell,m}\left[\left(\frac{\partial U_M}{\partial r} +\frac{U_L-U_M}{r}\right)\frac{\partial Y_\ell^m}{\partial\theta} + \left(\frac{\partial U_N}{\partial r} - \frac{U_N}{r}\right)\frac{1}{\sin\theta}\frac{\partial Y_\ell^m}{\partial\phi}\right].\end{align}$$

Comparing terms in the expansion, we see that
$$T_M=\mu\left[\frac{\partial U_M}{\partial r} + \frac{U_L-U_M}{r}\right]$$
and
$$T_N=\mu\left[\frac{\partial U_N}{\partial r} - \frac{U_N}{r}\right].$$
We would get the same results by expanding $\tau_{r\phi}$. There are two things of note here. First, these again give first order differential equations for $U_M$ and $U_N$. Second, the $\mathbf{N}$ component does not dependend on the $\mathbf{L}$ or $\mathbf{M}$ components, and vise-versa. We will see that for the equations of motion the toroidal and poloidal elements are decoupled as well.


### Gravity perturbations
Next, expand the perturbation to the gravitational potential
$$\nabla^2\psi = -4\pi G\nabla\cdot(\rho\mathbf{u}).$$
First, decompose the scalar field $\psi=\sum\Psi Y_\ell^m$ and expand the left-hand side:
$$\begin{align}
\nabla^2\psi &= \frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2\frac{\partial \psi}{\partial r}\right) + \nabla_S^2\psi\\
&= \sum_{\ell,m}\left[\frac{\partial^2 \Psi}{\partial r^2} + \frac{2}{r}\frac{\partial \Psi}{\partial r} -\frac{\ell(\ell+1)}{r^2}\Psi\right]Y_\ell^m.
\end{align}$$

To do so, expand the divergnce in the vector spherical harmonic basis
$$\begin{align}
\nabla\cdot\mathbf{u}&=\frac{1}{r^2}\frac{\partial}{\partial r}(r^2u_r) + \frac{1}{r\sin\theta}\frac{\partial}{\partial\theta}(u_\theta\sin\theta) + \frac{1}{r\sin\theta}\frac{\partial u_\phi}{\partial\phi}\\
&= \sum_{\ell,m}\left[\frac{2U_L}{r}+\frac{\partial U_L}{\partial r} - \frac{\ell(\ell-1)}{r}U_M\right]Y_\ell^m.
\end{align}$$

Thus, comparing terms in the expansion,
$$\frac{\partial^2 \Psi}{\partial r^2} + \frac{2}{r}\frac{\partial \Psi}{\partial r} -\frac{\ell(\ell+1)}{r^2}\Psi = -4\pi G\left[U_L\frac{\partial\rho}{\partial r} + \rho\left(\frac{2U_L}{r}+\frac{\partial U_L}{\partial r} - \frac{\ell(\ell-1)}{r}U_M\right)\right].$$

This is unsatisfying, though, because it is not a first-order differential equation. To convert it into coupled, first-order ODEs, we need to introduce a new variable. There are several such choices. We could choose (in spherical harmonic expansion) $Q_1=\frac{\partial \Psi}{\partial r}$ or alternatively $Q_2=\frac{\partial \Psi}{\partial r}+4\pi G\rho U_L + \frac{\ell+1}{r}\Psi$.

The first of these choices results in the following coupled equations
$$\begin{align}
\frac{\partial \Psi}{\partial r} &= Q_1\\
\frac{\partial Q_1}{\partial r} &= -4\pi G \left(\frac{\partial\rho}{\partial r} + \frac{4\mu\rho}{r(\lambda + 2\mu)}\right)U_L  +8\pi G \frac{\ell(\ell+1)\mu\rho}{r(\lambda + 2\mu)}U_M-4\pi G \frac{\rho}{\lambda + 2\mu}T_L + \frac{\ell(\ell+1)}{r^2}\Psi - \frac{2}{r}Q.
\end{align}$$

The second choice results in
$$\begin{align}
\frac{\partial \Psi}{\partial r} &= - 4\pi G U_L - \frac{\ell+1}{r}\Psi - Q_2\\
\frac{\partial Q_2}{\partial r} &= -4\pi G \rho\frac{\ell+1}{r}U_L + 4\pi G\rho\frac{\ell(\ell+1)}{r}U_M +\frac{\ell-1}{r}Q_2.
\end{align}$$
One of the benefits of this second definition for $Q$, which is sometimes called the _Generalized Flux_, is that it removes the explicit dependence of the equations on the density gradient $\frac{\partial \rho}{\partial r}$.

## Set up the integration
### Relaxation method
We could start with solutions that satisfy the lower boundary conditions, which we integrate to the surface independently and then combine in the right proportion to satisfy the upper boundary. This is called a _shooting method_. Alternatively, we can convert our equations into a difference equation and treat each neighboring pairs of point along our points as giving boundary and continuity conditions and relax them all into place simultaneously, what's called a _relaxation method_. 

One drawback of the relaxation method is that it requries a first guess at the solution - if it starts too far away, it may not converge to the correct solution very quickly. However, this drawback is only incurred once. We'll be doing this integration through the mantle many times (once for each order of the spherical expansion) and the solution doesn't change much between each order number. One iteration of relaxation ought to be sufficient to change the solutions from one order number to the next, a huge improvement on starting at the Core Mantle Boundary and shooting the differential equations again from the beginning.

At the heart of the relaxation method is the substitution of the differential equations with finite difference equations. Thus, our differential equation
$$\frac{d\mathbf{y}}{dx} = \mathbf{g}(x,\mathbf{y})$$
is approximated by
$$
\mathbf{0} = \mathbf{E_k} \equiv (\mathbf{y}_k - \mathbf{y}_{k-1}) - (x_k - x_{k-1})\,\mathbf{g}\left(\frac{1}{2}(x_k + x_{k-1}),\frac{1}{2}(\mathbf{y}_k + \mathbf{y}_{k-1})\right)\;\;\;\; k=1,\ldots,M-1
$$
on a grid of $M$ points. For our set of $N$ coupled differential equations, the equations above number $N(M-1)$. The remaining $N$ equations come from the boundary conditions.

There are $N_1$ independant equations at the first boundary
$$0 = \mathbf{E}_0\equiv \mathbf{B}(x_0,\mathbf{y}_0)$$
and $N_2\equiv N-N_1$ independent equations at the second
$$0 = \mathbf{E}_M\equiv \mathbf{C}(x_{M-1},\mathbf{y}_{M-1})$$

If the equation is in the linear form
$$\frac{d\mathbf{y}}{dx} = \mathbf{A}(x)\cdot\mathbf{y} + \mathbf{b}(x)$$
$$
\mathbf{0} = \mathbf{E_k} \equiv (\mathbf{y}_k - \mathbf{y}_{k-1}) - (x_k - x_{k-1})\,\left(\frac{1}{2}\mathbf{A}\left(\frac{1}{2}(x_k + x_{k-1})\right)\cdot(\mathbf{y}_k + \mathbf{y}_{k-1}) - \mathbf{b}\left(\frac{1}{2}(x_k + x_{k-1})\right)\right)
$$

The solution is a set of $y_{j,k}$ (as a reminder, these are the values of the $N$ variables $y_j$ at $M$ points $x_k$). So, given an initial $y_{j,k}$, we need an algorithm to find a $\Delta y_{j,k}$ such that $y_{j,k}+\Delta y_{j,k}$ is a better solution. To do this, Taylor expand the Finite Difference Equations for a small step in the solution
$$\mathbf{E}_k(\mathbf{y}_k + \Delta\mathbf{y}_k, \mathbf{y}_{k-1} + \Delta\mathbf{y}_{k-1}) \approx \mathbf{E}_k(\mathbf{y}_k,\mathbf{y}_{k-1}) + \sum_{n=0}^{N-1}\frac{\partial \mathbf{E}_k}{\partial y_{n,k}}\Delta y_{n,k} + \sum_{n=0}^{N-1}\frac{\partial \mathbf{E}_k}{\partial y_{n,k-1}}\Delta y_{n,k-1} = \mathbf{0},$$
set the whole thing to zero (for the optimal solution), and solve for the step. Given our linear differntial operator, this is very straightforward to write out, as
$$\begin{align}
\frac{\partial E_{j,k}}{\partial y_{n,k}} &= \delta_{jn} - \frac{1}{2}(x_k-x_{k-1})A_{jn}\left(\frac{1}{2}(x_k + x_{k-1})\right)\\
\frac{\partial E_{j,k}}{\partial y_{n,{k-1}}} &= -\delta_{jn} - \frac{1}{2}(x_k-x_{k-1})A_{jn}\left(\frac{1}{2}(x_k + x_{k-1})\right)
\end{align}$$

```python
def propMatElas(zarray, n, params, Q=1):
    """Generate the propagator matrix at all points in zarray. Should have
    shape (len(zarray), 6, 6)
    
    """
    # Check for individual z call
    zarray = np.asarray(zarray)
    singz = False
    if zarray.shape == ():
        zarray = zarray[np.newaxis]
        singz = True
        
        
    r = params.norms['r']
    muStar = params.norms['mu']
    G = params.G
    
    parvals = params.getParams(zarray/r)
    lam = muStar*parvals['bulk']
    mu = muStar*parvals['shear']
    rho = parvals['den']
    g = parvals['grav']
    grad_rho = np.gradient(parvals['den'])/np.gradient(zarray)#parvals['dend']/r
    #grad_rho_na = parvals['nonad']

    beta_i = 1./(lam+2*mu)
    gamma = mu*(3*lam+2*mu)*beta_i

    z_i = 1./zarray
    
    delsq = -n*(n+1)

    a = np.zeros((len(zarray), 6, 6))

    for i in range(len(zarray)):
        
        # r dU_L/dr
        a[i,0,0] = -2*lam[i]*beta_i[i]
        a[i,0,1] = -lam[i]*beta_i[i]*delsq
        a[i,0,2] = beta_i[i]*zarray[i]
        
        # r dU_M/dr
        a[i,1,0] = -1.
        a[i,1,1] = 1.
        a[i,1,2] = 0.
        a[i,1,3] = zarray[i]/mu[i]
        
        # r dT_L/dr
        if Q == 1:
            a[i,2,0] = 4*(gamma[i]*z_i[i] - rho[i]*g[i]) + G*rho[i]**2
        else:
            a[i,2,0] = 4*(gamma[i]*z_i[i] - rho[i]*g[i])
        a[i,2,1] = (2*gamma[i]*z_i[i]-rho[i]*g[i])*delsq
        a[i,2,2] = -4*mu[i]*beta_i[i]
        a[i,2,3] = -delsq
        if Q == 1:
            a[i,2,4] = 0
        else:
            a[i,2,4] = -(n+1)*rho[i]
        a[i,2,5] = zarray[i]*rho[i]
        
        # r dT_M/dr
        a[i,3,0] = -2*gamma[i]*z_i[i]+rho[i]*g[i]
        a[i,3,1] = -2*mu[i]*(2*delsq*(lam[i]+mu[i])*beta_i[i] + 1)*z_i[i]
        a[i,3,2] = -lam[i]*beta_i[i]
        a[i,3,3] = -3
        a[i,3,4] = rho[i]
        
        # r dPsi/dr 
        if Q == 1:
            a[i,4,0] = 0.
        else: 
            a[i,4,0] = -G*rho[i]*zarray[i]
        a[i,4,1] = 0.
        a[i,4,2] = 0.
        a[i,4,3] = 0.
        if Q == 1:
            a[i,4,4] = 0.
        else:
            a[i,4,4] = -(n+1)
        a[i,4,5] = zarray[i]
        
        # r dQ/dr
        if Q == 1:
            a[i,5,0] = -G*(4*rho[i]*mu[i]*beta_i[i]+zarray[i]*grad_rho[i])
            a[i,5,1] = -2*delsq*G*rho[i]*mu[i]*beta_i[i]
            a[i,5,2] = -G*zarray[i]*rho[i]*beta_i[i]
            a[i,5,3] = 0.
            a[i,5,4] = -delsq*z_i[i]
            a[i,5,5] = -2
        else:
            a[i,5,0] = -G*rho[i]*(n+1)
            a[i,5,1] = -delsq*G*rho[i]
            a[i,5,2] = 0
            a[i,5,3] = 0.
            a[i,5,4] = 0
            a[i,5,5] = n-1

    if singz:
        return z_i*a[0]
    else:
        return (z_i*a.T).T

class SphericalElasSMat(object):
    """Class that provides smatrix to Solvde for Elastic solutions."""
    def __init__(self, n, z, params, Q=1):
        self.n = n
        self.z = z
        self.params = params
        self.Q = Q

        self.mpt = len(self.z)
        self.updateProps()
        
    def updateProps(self, n=None, z=None):
        self.n = n or self.n
        self.z = self.z if z is None else z
        zmids = 0.5*(self.z[1:]+self.z[:-1])
        self.A = propMatElas(zmids, self.n, self.params, self.Q)

    def smatrix(self, k, k1, k2, jsf, is1, isf, indexv, s, y):
        Q = self.Q
        if k == k1:      # Core-Mantle boundary conditions.
            rCore = self.params.rCore
            paramsCore = self.params(rCore)
            rhoCore = paramsCore['den']
            gCore = paramsCore['grav']
            
            denCore = self.params.denCore
            difden = denCore-rhoCore
            G = self.params.G

            rstar = self.params.norms['r']
            mustar = self.params.norms['mu']
            
            # Radial stress on the core.
            if Q == 1:
                s[3, 6+indexv[0]] = - 0.33*G*rstar*rCore*denCore**2#-gCore*denCore
            else:
                s[3, 6+indexv[0]] = -0.33*G*rstar*rCore*denCore**2
            s[3, 6+indexv[1]] = 0.
            s[3, 6+indexv[2]] = 1.
            s[3, 6+indexv[3]] = 0.
            s[3, 6+indexv[4]] = -denCore    
            s[3, 6+indexv[5]] = 0.
            if Q == 1:
                s[3, jsf] = (y[2,0] - 0.33*G*rstar*rCore*denCore**2*y[0,0] - #gCore*denCore*y[0,0] - 
                            denCore*y[4,0])
            else:
                s[3, jsf] = (y[2,0] - 0.33*G*rstar*rCore*denCore**2*y[0,0] - 
                            denCore*y[4,0])
                                            
            # Poloidal stress on the core.
            s[4, 6+indexv[0]] = 0.          
            s[4, 6+indexv[1]] = 0.          
            s[4, 6+indexv[2]] = 0.          
            s[4, 6+indexv[3]] = 1.          
            s[4, 6+indexv[4]] = 0.          
            s[4, 6+indexv[5]] = 0.          
            s[4, jsf] = y[3,0]

            # gravitational potential perturbation on core.
            if Q == 1:
                s[5, 6+indexv[0]] = -G*difden
                s[5, 6+indexv[1]] = 0.
                s[5, 6+indexv[2]] = 0.
                s[5, 6+indexv[3]] = 0.
                s[5, 6+indexv[4]] = -(self.n)/rCore/rstar
                s[5, 6+indexv[5]] = 1.
                s[5, jsf] = (y[5,0] - G*difden*y[0,0] - self.n/rCore/rstar*y[4,0])
            else:
                s[5, 6+indexv[0]] = -G*denCore
                s[5, 6+indexv[1]] = 0.
                s[5, 6+indexv[2]] = 0.#-3/rCore/rstar/rhoCore
                s[5, 6+indexv[3]] = 0
                s[5, 6+indexv[4]] = -(2*self.n+1)/rCore/rstar#-2*(self.n-1)/rCore/rstar
                s[5, 6+indexv[5]] = 1.
                s[5, jsf] = (y[5,0] - G*denCore*y[0,0] - (2*self.n+1)/rCore/rstar*y[4,0])
                #s[5, jsf] = (y[5,0] - 3/rCore/rstar/rhoCore*y[2,0] - 2*(self.n-1)/rCore/rstar*y[4,0])
 
                

        elif k >= k2:     # Surface boundary conditions.
            paramsSurf = self.params(1.)
            gSurf = paramsSurf['grav']
            rhoSurf = paramsSurf['den']
            G = self.params.G
            

            rstar = self.params.norms['r']
            mustar = self.params.norms['mu']

            # Radial stress on surface.
            s[0, 6+indexv[0]] = 0.
            s[0, 6+indexv[1]] = 0.
            s[0, 6+indexv[2]] = 1.
            s[0, 6+indexv[3]] = 0.
            s[0, 6+indexv[4]] = 0.
            s[0, 6+indexv[5]] = 0.
            s[0, jsf] = (y[2, self.mpt-1] + gSurf)

            # Poloidal stress on surface.
            s[1, 6+indexv[0]] = 0.
            s[1, 6+indexv[1]] = 0.
            s[1, 6+indexv[2]] = 0.    
            s[1, 6+indexv[3]] = 1.    
            s[1, 6+indexv[4]] = 0.    
            s[1, 6+indexv[5]] = 0.    
            s[1, jsf] = y[3, self.mpt-1]
                                      
            # gravitational potential perturbation on surface.
            if Q == 1:
                s[2, 6+indexv[0]] = G*rhoSurf
            else:
                s[2, 6+indexv[0]] = 0.
            s[2, 6+indexv[1]] = 0.    
            s[2, 6+indexv[2]] = 0.    
            s[2, 6+indexv[3]] = 0.
            if Q == 1:
                s[2, 6+indexv[4]] = (self.n+1)/rstar
            else:
                s[2, 6+indexv[4]] = 0.
            s[2, 6+indexv[5]] = 1.
            if Q == 1:
                s[2, jsf] = (y[5, self.mpt-1] + G +
                                 G*rhoSurf*y[0,self.mpt-1]+
                                 (self.n+1)*y[4,self.mpt-1]/rstar)
            else:
                s[2, jsf] = y[5, self.mpt-1] + G

        else:           # Finite differences.
            zsep = (self.z[k] - self.z[k-1])
            A = 0.5*zsep*self.A[k-1]
            tmp = np.zeros_like(self.z)
            interior_smatrix_fast(6, k, jsf, A, tmp, y, indexv, s) 
                      
        return s

def interior_smatrix_fast(n, k, jsf, A, b, y, indexv, s):
    for i in range(n):
        rgt = 0.
        for j in range(n):
            if i==j:
                s[i, indexv[j]]   = -1. - A[i,j]
                s[i, n+indexv[j]] =  1. - A[i,j]
            else:
                s[i, indexv[j]]   = -A[i,j]
                s[i, n+indexv[j]] = -A[i,j]
            rgt += A[i,j] * (y[j, k] + y[j, k-1])
        s[i, jsf] = y[i, k] - y[i, k-1] - rgt - b[i]
```

## Normalized equations

```python
def propMatElas(zarray, n, params):
    """Generate the propagator matrix at all points in zarray. Should have
    shape (len(zarray), 6, 6)
    
    """
    # Check for individual z call
    zarray = np.asarray(zarray)
    singz = False
    if zarray.shape == ():
        zarray = zarray[np.newaxis]
        singz = True
        
        
    r = params.norms['r']
    muStar = params.norms['mu']
    G = params.G
    
    parvals = params.getParams(zarray)
    lam = muStar*parvals['bulk']
    mu = muStar*parvals['shear']
    rho = parvals['den']
    g = parvals['grav']
    
    g0 = params.getParams(1.)['grav']
    re = params.norms['r']
    rhobar = g0/params.G/re
    
    mu = mu/rhobar/re/g0
    lam = lam/rhobar/re/g0
    g = g/g0
    rho = rho/rhobar
    
    #grad_rho_na = parvals['nonad']

    beta_i = 1./(lam+2*mu)
    gamma = mu*(3*lam+2*mu)*beta_i

    z_i = 1./zarray

    a = np.zeros((len(zarray), 6, 6))

    for i in range(len(zarray)):
        
        # r dh/dr
        a[i,0,0] = -2*lam[i]*beta_i[i]
        a[i,0,1] = lam[i]*beta_i[i]*(n+1)
        a[i,0,2] = beta_i[i]*zarray[i]*(2*n+1)
        
        # r dL/dr
        a[i,1,0] = -n
        a[i,1,1] = 1.
        a[i,1,2] = 0.
        a[i,1,3] = zarray[i]/mu[i]*(2*n+1)
        
        # r df_L/dr
        a[i,2,0] = 4*(gamma[i]*z_i[i] - rho[i]*g[i])/(2*n+1)
        a[i,2,1] = -(2*gamma[i]*z_i[i] - rho[i]*g[i])*(n+1)/(2*n+1)
        a[i,2,2] = -4*mu[i]*beta_i[i]
        a[i,2,3] = (n+1)
        a[i,2,4] = -rho[i]*(n+1)/(2*n+1)
        a[i,2,5] = zarray[i]*rho[i]
        
        # r dF_M/dr
        a[i,3,0] = (rho[i]*g[i]-2*gamma[i]*z_i[i])*n/(2*n+1)
        a[i,3,1] = 2*mu[i]*z_i[i]*(2*n*(n+1)*(lam[i]+mu[i])*beta_i[i] - 1)/(2*n+1)
        a[i,3,2] = -lam[i]*beta_i[i]*n
        a[i,3,3] = -3
        a[i,3,4] = rho[i]/(2*n+1)*n
        
        # r dk_d/dr 
        a[i,4,0] = -rho[i]*zarray[i]
        a[i,4,1] = 0.
        a[i,4,2] = 0.
        a[i,4,3] = 0.
        a[i,4,4] = -(n+1)
        a[i,4,5] = zarray[i]*(2*n+1)
        
        # r dQ/dr
        a[i,5,0] = -rho[i]*(n+1)/(2*n+1)
        a[i,5,1] = rho[i]*(n+1)/(2*n+1)
        a[i,5,2] = 0
        a[i,5,3] = 0.
        a[i,5,4] = 0
        a[i,5,5] = n-1

    if singz:
        return z_i*a[0]
    else:
        return (z_i*a.T).T

class SphericalElasSMat(object):
    """Class that provides smatrix to Solvde for Elastic solutions."""
    def __init__(self, n, z, params):
        self.n = n
        self.z = z
        self.params = params

        self.mpt = len(self.z)
        self.updateProps()
        
    def updateProps(self, n=None, z=None):
        self.n = n or self.n
        self.z = self.z if z is None else z
        zmids = 0.5*(self.z[1:]+self.z[:-1])
        self.A = propMatElas(zmids, self.n, self.params)

    def smatrix(self, k, k1, k2, jsf, is1, isf, indexv, s, y):
        
        paramSurf = self.params(1.)
        rhobar = paramSurf['grav']/self.params.G/self.params.norms['r']
        
        if k == k1:      # Core-Mantle boundary conditions.
            
            g0 = self.params.getParams(1.)['grav']
            re = self.params.norms['r']
            rhobar = g0/self.params.G/re
            
            rCore = self.params.rCore
            paramsCore = self.params(rCore)
            gCore = paramsCore['grav']/g0
            
            denCore = self.params.denCore/rhobar
            
            
            # Radial stress on the core.
            s[3, 6+indexv[0]] = -0.33*rCore*denCore**2/(2*self.n+1)
            s[3, 6+indexv[1]] = 0.
            s[3, 6+indexv[2]] = 1.
            s[3, 6+indexv[3]] = 0.
            s[3, 6+indexv[4]] = -denCore/(2*self.n+1)
            s[3, 6+indexv[5]] = 0.
            s[3, jsf] = (y[2,0] - 0.33*rCore*denCore**2/(2*self.n+1)*y[0,0] - 
                            denCore/(2*self.n+1)*y[4,0])
                                            
                                            
            # Poloidal stress on the core.
            s[4, 6+indexv[0]] = 0.          
            s[4, 6+indexv[1]] = 0.          
            s[4, 6+indexv[2]] = 0.          
            s[4, 6+indexv[3]] = 1.          
            s[4, 6+indexv[4]] = 0.          
            s[4, 6+indexv[5]] = 0.          
            s[4, jsf] = y[3,0]

            # gravitational potential perturbation on core.
            s[5, 6+indexv[0]] = -denCore/(2*self.n+1)
            s[5, 6+indexv[1]] = 0.
            s[5, 6+indexv[2]] = 0.
            s[5, 6+indexv[3]] = 0
            s[5, 6+indexv[4]] = -1/rCore
            s[5, 6+indexv[5]] = 1.
            s[5, jsf] = (y[5,0] - denCore/(2*self.n+1)*y[0,0] - 1/rCore*y[4,0])
 
                

        elif k >= k2:     # Surface boundary conditions.
            g0 = self.params.getParams(1.)['grav']
            re = self.params.norms['r']
            rhobar = g0/self.params.G/re
            
            paramsSurf = self.params(1.)
            gSurf = paramsSurf['grav']
            rhoSurf = paramsSurf['den']
            G = self.params.G
            

            rstar = self.params.norms['r']
            mustar = self.params.norms['mu']

            # Radial stress on surface.
            s[0, 6+indexv[0]] = 0.
            s[0, 6+indexv[1]] = 0.
            s[0, 6+indexv[2]] = 1.
            s[0, 6+indexv[3]] = 0.
            s[0, 6+indexv[4]] = 0.
            s[0, 6+indexv[5]] = 0.
            s[0, jsf] = (y[2, self.mpt-1] + 1)

            # Poloidal stress on surface.
            s[1, 6+indexv[0]] = 0.
            s[1, 6+indexv[1]] = 0.
            s[1, 6+indexv[2]] = 0.    
            s[1, 6+indexv[3]] = 1.    
            s[1, 6+indexv[4]] = 0.    
            s[1, 6+indexv[5]] = 0.    
            s[1, jsf] = y[3, self.mpt-1]
                                      
            # gravitational potential perturbation on surface.
            s[2, 6+indexv[0]] = 0.
            s[2, 6+indexv[1]] = 0.    
            s[2, 6+indexv[2]] = 0.    
            s[2, 6+indexv[3]] = 0.
            s[2, 6+indexv[4]] = 0.
            s[2, 6+indexv[5]] = 1.
            s[2, jsf] = y[5, self.mpt-1] + 1

        else:           # Finite differences.
            zsep = (self.z[k] - self.z[k-1])
            A = 0.5*zsep*self.A[k-1]
            tmp = np.zeros_like(self.z)
            interior_smatrix_fast(6, k, jsf, A, tmp, y, indexv, s) 
                      
        return s

def interior_smatrix_fast(n, k, jsf, A, b, y, indexv, s):
    for i in range(n):
        rgt = 0.
        for j in range(n):
            if i==j:
                s[i, indexv[j]]   = -1. - A[i,j]
                s[i, n+indexv[j]] =  1. - A[i,j]
            else:
                s[i, indexv[j]]   = -A[i,j]
                s[i, n+indexv[j]] = -A[i,j]
            rgt += A[i,j] * (y[j, k] + y[j, k-1])
        s[i, jsf] = y[i, k] - y[i, k-1] - rgt - b[i]
```

```python
class SphericalElasSMatAM(object):
    """Class that provides smatrix to Solvde for Elastic solutions."""
    def __init__(self, n, z, params, delta):
        self.n = n
        self.z = z
        self.params = params
        self.delta = delta

        self.mpt = len(self.z)
        self.updateProps()
        
    def updateProps(self, n=None, z=None):
        if n is None: n = self.n
        self.n = n
        if z is None: z = self.z
        self.z = z
        zmids = 0.5*(self.z[1:]+self.z[:-1])
        self.A = propMatElas(zmids, self.n, self.params)

    def smatrix(self, k, k1, k2, jsf, is1, isf, indexv, s, y):
        if k == k1:      # Core-Mantle boundary conditions.
            self.updateProps(z=y[-1,:])
            
            g0 = self.params.getParams(1.)['grav']
            re = self.params.norms['r']
            rhobar = g0/self.params.G/re
            
            rCore = self.params.rCore
            paramsCore = self.params(rCore)
            gCore = paramsCore['grav']/g0
            
            denCore = self.params.denCore/rhobar

            s[4, 9+indexv[0]] = -0.33*rCore*denCore**2/(2*self.n+1)
            s[4, 9+indexv[1]] = 0.
            s[4, 9+indexv[2]] = 1.
            s[4, 9+indexv[3]] = 0.
            s[4, 9+indexv[4]] = -denCore/(2*self.n+1)   
            s[4, 9+indexv[5]] = 0.
            s[4, 9+indexv[6]] = 0.
            s[4, 9+indexv[7]] = 0.    
            s[4, 9+indexv[8]] = 0.
            s[4, jsf] = (y[2,0] - 0.33*rCore*denCore**2/(2*self.n+1)*y[0,0] - 
                            denCore/(2*self.n+1)*y[4,0])
                                            
            # Poloidal stress on the core.
            s[5, 9+indexv[0]] = 0.
            s[5, 9+indexv[1]] = 0.
            s[5, 9+indexv[2]] = 0.
            s[5, 9+indexv[3]] = 1.
            s[5, 9+indexv[4]] = 0.
            s[5, 9+indexv[5]] = 0.
            s[5, 9+indexv[6]] = 0.
            s[5, 9+indexv[7]] = 0.
            s[5, 9+indexv[8]] = 0.
            s[5, jsf] = y[3,0]

            # gravitational potential perturbation on core.
            s[6, 9+indexv[0]] = -denCore/(2*self.n+1)
            s[6, 9+indexv[1]] = 0.
            s[6, 9+indexv[2]] = 0.
            s[6, 9+indexv[3]] = 0.
            s[6, 9+indexv[4]] = -1/rCore
            s[6, 9+indexv[5]] = 1.
            s[6, 9+indexv[6]] = 0.
            s[6, 9+indexv[7]] = 0.
            s[6, 9+indexv[8]] = 0.
            s[6, jsf] = (y[5,0] - denCore/(2*self.n+1)*y[0,0] - 1/rCore*y[4,0])

            # core boundary r value.
            s[7, 9+indexv[0]] = 0.
            s[7, 9+indexv[1]] = 0.
            s[7, 9+indexv[2]] = 0.
            s[7, 9+indexv[3]] = 0.
            s[7, 9+indexv[4]] = 0.
            s[7, 9+indexv[5]] = 0.
            s[7, 9+indexv[6]] = 0.
            s[7, 9+indexv[7]] = 0.
            s[7, 9+indexv[8]] = 1.
            s[7,jsf] = y[8, 0] - rCore
            
            # core boundary r-point density value.
            s[8, 9+indexv[0]] = 0.
            s[8, 9+indexv[1]] = 0.
            s[8, 9+indexv[2]] = 0.
            s[8, 9+indexv[3]] = 0.
            s[8, 9+indexv[4]] = 0.
            s[8, 9+indexv[5]] = 0.
            s[8, 9+indexv[6]] = 1.
            s[8, 9+indexv[7]] = 0.
            s[8, 9+indexv[8]] = 0.
            s[8,jsf] = y[6, 0]
            
        elif k >= k2:     # Surface boundary conditions.
            g0 = self.params.getParams(1.)['grav']
            re = self.params.norms['r']
            rhobar = g0/self.params.G/re
            
            paramsSurf = self.params(1.)
            gSurf = paramsSurf['grav']
            rhoSurf = paramsSurf['den']
            G = self.params.G
            

            rstar = self.params.norms['r']
            mustar = self.params.norms['mu']

            # Radial stress on surface.
            s[0, 9+indexv[0]] = 0.
            s[0, 9+indexv[1]] = 0.
            s[0, 9+indexv[2]] = 1.
            s[0, 9+indexv[3]] = 0.
            s[0, 9+indexv[4]] = 0.
            s[0, 9+indexv[5]] = 0.
            s[0, 9+indexv[6]] = 0.
            s[0, 9+indexv[7]] = 0.
            s[0, 9+indexv[8]] = 0.
            s[0, jsf] = (y[2, self.mpt-1] + 1)

            # Poloidal stress on surface.
            s[1, 9+indexv[0]] = 0.
            s[1, 9+indexv[1]] = 0.
            s[1, 9+indexv[2]] = 0.    
            s[1, 9+indexv[3]] = 1.    
            s[1, 9+indexv[4]] = 0.    
            s[1, 9+indexv[5]] = 0.
            s[1, 9+indexv[6]] = 0.    
            s[1, 9+indexv[7]] = 0.    
            s[1, 9+indexv[8]] = 0.
            s[1, jsf] = y[3, self.mpt-1]
                                      
            # gravitational potential perturbation on surface.
            s[2, 9+indexv[0]] = 0.
            s[2, 9+indexv[1]] = 0.    
            s[2, 9+indexv[2]] = 0.    
            s[2, 9+indexv[3]] = 0.    
            s[2, 9+indexv[4]] = 0.
            s[2, 9+indexv[5]] = 1.
            s[2, 9+indexv[6]] = 0.    
            s[2, 9+indexv[7]] = 0.    
            s[2, 9+indexv[8]] = 0.
            s[2, jsf] = (y[5, self.mpt-1] + 1)
            
            # surface boundary r poisition
            s[3, 9+indexv[0]] = 0.
            s[3, 9+indexv[1]] = 0.
            s[3, 9+indexv[2]] = 0.
            s[3, 9+indexv[3]] = 0.
            s[3, 9+indexv[4]] = 0.
            s[3, 9+indexv[5]] = 0.
            s[3, 9+indexv[6]] = 0.
            s[3, 9+indexv[7]] = 0.
            s[3, 9+indexv[8]] = 1.
            s[3,jsf] = y[8, self.mpt-1] - 1

        else:           # Finite differences.
            zsep = (self.z[k] - self.z[k-1])
            A = self.A[k-1]
            Anew = np.zeros((9,9))
            
            #dydx = np.abs(self.A[k-1].dot(y[:6,k-1])*rstar)
            #dydx = np.abs(self.A[:,0,:].dot(y[:6,:-1]))
            #dydx = np.array([self.A[i,0,:].dot(0.5*y[:6,i] + 0.5*y[:6,i+1]) for i in range(y.shape[1]-1)])
            #dydx = np.abs(dydx/y[0,:-1])
            #dydx = np.abs(dydx)
            
            #dydx = np.sum(np.minimum(np.abs(A.dot(y[:6,k-1])/np.maximum(np.abs(y[:6,k-1]),1e-5)), 10))/6.
            dydx = np.sum(np.minimum(np.abs(A.dot(y[:6,k-1])), 10))/6.
            
            #delta = (1-self.params.rCore)/(y.shape[1] - 1)/100.
            
            #phi = 1/delta*(0.9 + 0.1*np.abs(dydx[0]/y[0,k-1]))
            #print dydx[0]
            phi = 1./self.delta*(.5 + .5*dydx)
            #phi = 1/delta*(0.01 + 0.99*dydx[k-1]/np.max(dydx))
            
            Anew[:6,:6]=A
            Anew[-3,-2]=1.
            Anew[-1,-2]=1./phi
            
            Anew[:6] *= y[-2,k-1]/phi
            
            Anew = 0.5*Anew
            
            tmp = np.zeros_like(self.z)
            interior_smatrix_fast(9, k, jsf, Anew, tmp, y, indexv, s) 
                      
        return s
```
